```python
#what we done is we grouped the Airline and Total stops and we found the median of those ..
#then kept those grouped values in a dictionary format
dict1=final_df.groupby(by=['Airline','Total_Stops']).median()['Price'].to_dict()
dict1
```

```python
#here we set the index as Airline and Total stops for adding gropued median values
df=final_df.set_index(['Airline','Total_Stops'])
```

```python
#using this code we fill the price for each NaN values, 
#replaced for  row values so that we given axis =0
# and finally we reset the index 
df['Price'].fillna(dict1,inplace=True,axis=0)
df=df.reset_index()
df
```

```python
#again there is one null value in price so we filled that value with mean
df['Price']=df['Price'].fillna(df['Price'].mean())
```

# MultipleLinear Regression

Now we had completely cleaned the data for further analysis\
Now we had changed all the variables to numeric for analysis

```python
X=df.iloc[:,df.columns!='Price']
X
#X shows our independent variable
#here we located that, alll the variables without price variable
```

```python
y=df.loc[:,'Price']
y
#this is our dependet variable
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
regr.predict
```

```python
from sklearn import metrics
import numpy as np
print('y-intercept             :' , regr.intercept_)
print('beta coefficients       :' , regr.coef_)
```

```python
# Splitting the data into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
regr.fit(X_train, y_train)
```

```python
Y_pred = regr.predict(X_test)
Y_pred
```

```python
print('Mean Abs Error MAE      :' ,metrics.mean_absolute_error(y_test,Y_pred))
print('Mean Sqrt Error MSE     :' ,metrics.mean_squared_error(y_test,Y_pred))
print('Root Mean Sqrt Error RMSE:' ,np.sqrt(metrics.mean_squared_error(y_test,Y_pred)))
print('r2 value                :' ,metrics.r2_score(y_test,Y_pred))
```
